--------------------<<<<<<<<<<<<<<<<<<<<<<<SelfDefine scale for vanderSentiment>>>>>>>>>>>>>>>>>>>-----------------------------

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
def sentiment_analyzer_scores(text):
    score = analyzer.polarity_scores(text)
    lb = score['compound']
#[-1 to 0]: negative,[0]: neutral[0 to +1]: positive
    if lb >= 0.05:
        return 1
    elif (lb > -0.05) and (lb < 0.05):
        return 0
    else:
        return -1
#.............
#.............funcation call
Ra=sentiment_analyzer_scores("The food here is soso")
print(Ra)
Rb=sentiment_analyzer_scores("The food here is delicious")
print(Rb)
Rc=sentiment_analyzer_scores("The food here is not delicious")
print(Rc)
```
